```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df.isna().sum()
```

```python
data = df.copy()
data['MEDV'] = value.copy()['MEDV'] # combining input and output into one table
train_data = data.sample(frac=0.8, random_state=0)
test_data = data.drop(train_data.index)
```

```python
# this can be done more efficiently (without copying) by using pop. I'm only doing
# it this way because if you run this box twice in a row when you use pop it causes
# an error which makes it hard to modify things
train_labels = train_data.copy()['MEDV']
test_labels = test_data.copy()['MEDV']
```

```python
train_data = train_data.drop('MEDV', axis=1)
test_data = test_data.drop('MEDV', axis=1)
```

```python
norm_test_data = (test_data - test_data.describe().transpose()['mean']) / test_data.describe().transpose()['std']
norm_train_data = (train_data - train_data.describe().transpose()['mean']) / train_data.describe().transpose()['std']
```

```python
def build_model():
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=[len(norm_train_data.keys())]),
        layers.Dense(2, activation='relu'),
        #layers.Dense(16, activation='relu'),
        layers.Dense(1)
    ])
    
    #optimizer = tf.keras.optimizers.RMSprop(0.001)

    model.compile(loss='mse',
                optimizer='adam',
                metrics=['mae', 'mse'])
    return model
```

```python
nn_model = build_model()
```

```python
nn_model.summary()
```

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')

EPOCHS = 500

history = nn_model.fit(
    norm_train_data, train_labels,
    epochs=EPOCHS, validation_split=0.2, verbose=0,
    callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch
    
    plt.figure()
    plt.title('Mean Absolute Error vs. Number of Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Absolute Error [MEDV]')
    plt.plot(hist['epoch'], hist['mae'],
            label='Train Error')
    plt.plot(hist['epoch'], hist['val_mae'],
            label='Val Error')
    plt.ylim([0,30])
    plt.legend()
    
    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error [MEDV^2]')
    plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
    plt.ylim([0,60])
    plt.legend()
    plt.show()
    
plot_history(history)  
```

```python
loss, mae, mse = nn_model.evaluate(norm_test_data, test_labels, verbose=2)
r2 = r2_score(test_labels, nn_model.predict(norm_test_data).flatten())

print("Testing set Mean Abs Error: {:5.2f}".format(mae))
print("Testing set Mean Squared Error: {:5.2f}".format(mse))
print(r2)
```

```python
test_predictions = nn_model.predict(norm_test_data).flatten()

#a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('Actual Median Values [in $1000s]')
plt.ylabel('Predicted Median Values [in $1000s]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
# Now try training model with early stop to get number of epochs correct
nn_model_es = build_model()

# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=30)

history_es = nn_model_es.fit(norm_train_data, train_labels, epochs=EPOCHS,
                    validation_split = 0.2, verbose=0, callbacks=[early_stop, PrintDot()])

plot_history(history_es)
```

```python
loss, mae, mse = nn_model_es.evaluate(norm_test_data, test_labels, verbose=2)
r2 = r2_score(test_labels, nn_model_es.predict(norm_test_data).flatten())

print("Testing set Mean Abs Error: {:5.2f}".format(mae))
print("Testing set Mean Squared Error: {:5.2f}".format(mse))
print(r2)
```

```python
test_predictions = nn_model.predict(norm_test_data).flatten()

#a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('Actual Median Values [in $1000s]')
plt.ylabel('Predicted Median Values [in $1000s]')
lims = [0, 50]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```
